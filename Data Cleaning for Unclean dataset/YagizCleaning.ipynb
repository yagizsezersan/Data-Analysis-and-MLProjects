#### c. Drop and transform the dataset

```python
# First we need to see how many null values we have as per column
df.isnull().sum()
```

```python
# As per df.(info) we have total 2004 rows of entry so Unnamed columns seems all null value.
df.drop('Unnamed: 7',axis=1,inplace=True)
```

```python
# Check again to be sure if related column is dropped
df.info()
df.isnull().sum()
```

```python
# Drop rows if all data are missing so that s why how=all parameter is used.
df.dropna(how='all',inplace=True)
```

```python
# After dropping, we still have 1970 rows of data
df.info()
df.isnull().sum()
```

```python
# COST2 has more null values than COST so COST2 should be dropped
# First CAD$ and EUR indication should be deleted

df.COST2 = df.COST2.str.replace("EUR","")
#df.loc[df.COST.str.contains('(?i)CAD$'), 'COST'] = ' '
#df.loc[df.COST2.str.contains('(?i)EUR'), 'COST2'] = ' '
```

```python
# Since below code does not delete CAD$ from string, I applied another workaround with next code
df.COST = df.COST.str.replace("CAD$","",regex=True)

#Later on I figured out regex=False should be used i/o True but I proceed with splitting to practice this way
df.head()
```

```python
# split COST column as CAD# and number 
df[['currency','COST1']] = df.COST.str.split("$",expand=True)
print(df,"\n")

# then drop both column to left COST1 column with only number.
df.drop(['currency','COST'],axis=1,inplace=True)
df.head()
```

```python
# change COST objects into float
df[['COST1','COST2']] = df[['COST1','COST2']].astype('float64')
df.info()
```

```python
# fill missing value in COST1-cad with COST2-eur via considering currency 1 CAD = 0.67 EUR
df.COST1.fillna(0.67*df.COST2, inplace=True)

#seems there is no null value remains for COST1 so we can drop now COST2
df.isnull().sum()
```

```python
# Drop COST2 to left only one data that represents cost
df.drop('COST2',axis=1,inplace=True)
df.info()
```

```python
# drop remaining nan/null values
df.dropna(inplace=True)
df.isnull().sum().sum()
```

```python
df.info()
#below code proves that there are no any null values anymore
df.isnull().sum()
```

```python
# before next process, let s take a look final situation of our data
df.head()
```

#### d. Organize the label of columns to have clear vision

```python
# Change whitespace with underscore
#df.columns.str.replace(" ","_")
df.rename(columns=lambda x: x.replace(" ", "_"), inplace=True)
df.info()
```

```python
# Transform header names into small case letters
#df.columns.str.lower()
df.rename(columns=lambda x: x.lower(), inplace=True)
df.head()
```

#### e. Remove special characters from the data and format some column data


```python
# Delete % from discount column
df.discount = df.discount.str.replace("%","")
print(df.discount)
```

```python
# Update duration column into minutes
# First let s remove h and m then split it
df.duration = df.duration.str.replace("h","").str.replace("m","").str.split(" ")

# cal_min() to calculate duration column in minute
def cal_min(x):
    if len(x) == 2:
        hours = float(x[0])
        minutes = float(x[1])
    else:
        hours = 0
        minutes = float(x[0])
    return hours * 60 + minutes

df.duration = df.duration.apply(cal_min)

print(df.duration)
```

```python
# Update last_updated as last 2 digits of year value
df['last_updated'] = df['last_updated'].apply(lambda x: x.split("/")[1][-2:])
print(df.last_updated)
```

```python
# change data type as float for discount, duration and last_updated columns
df[['discount','duration','last_updated']] = df[['discount','duration','last_updated']].astype('float64')
df.info()
```

```python
df.head()
```

#### f. Reduce the number of unique values in the category

```python
# unique values and count for category column
df.category.unique()
```

```python
# count of unique data for category column
df.category.value_counts()
```

```python
# Drop the rows which category value has less than 20
# Assign all categories except below ones to drop them
category_drop = ['finance accounting', 'other']  

df = df[~df['category'].isin(category_drop)]
```

```python
# To confirm if 'finance accounting', 'other' are dropped or not
df.category.value_counts()
```

```python
# Combine any category that contains the following word: 
# personal, music, health, lifestyle or photography into personal development category
df.loc[df['category'].str.contains('(?i)personal|music|health|lifestyle|photography'), 'category'] = 'personal development'
```

```python
# We have eight record for category now
print(df.category.value_counts(),"\n")

print(f"The category column has {df.category.nunique()} unique rows now")
```

```python
# Replace whitespace with underscore in category column
df.category = df.category.str.replace(" ","_")
df.category.unique()
```

```python
# Number of course for sub_category
df.sub_category.nunique()
```

#### g. Drop unnecessary columns and reset the index

```python
# Drop course_name and sub_category
df.drop(['course_name','sub_category'],axis=1,inplace=True)

# Reset the index and ensure that the previous index has been not kept
df.reset_index(drop=True, inplace=True)
```

```python
# Rename cost1 as cost for clear explanation of header
df.rename(columns={'cost1':'cost'}, inplace=True)

# Currently all numeric columns are indicated as float64 except category column
df.info()
```

#### h. Save the Cleaned Dataset 

```python
# Save updated dataset and indicate index as False since we do not want to include index columns
df.to_csv('cleaned.csv', index=False)
```
