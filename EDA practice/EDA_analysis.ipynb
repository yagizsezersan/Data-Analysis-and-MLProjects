```python
# Use variance threshold method to select features
# Separate features and target 
from sklearn.feature_selection import VarianceThreshold
y = df_dummy.avg_rating
x = df_dummy.drop('avg_rating', axis=1)
```

```python
# Apply the variance threshold method and get selected feature names
vt = VarianceThreshold(threshold=0.1)
vt.fit(x)
pd.DataFrame({'Variance': vt.variances_, 'select_status': vt.get_support()}, index=x.columns)
```

```python
# Save it to df_variance with the selected features
df_variance = x.iloc[:,vt.get_support()]
df_variance.head()
```

```python
# Use SelectKBest method with k=8
target = df_dummy.avg_rating
features = df_dummy.drop('avg_rating', axis=1)

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression # use f_regression for regression algorithm
from sklearn.model_selection import train_test_split

# Select 8 features
kbest = SelectKBest(score_func=f_regression, k=8)
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

kbest.fit_transform(X_train, y_train)
# Save it to df_selKBest with the selected features
df_selKBest = features.iloc[:,kbest.get_support()]
df_selKBest.head()
```
